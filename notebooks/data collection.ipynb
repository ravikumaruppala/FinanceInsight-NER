```python
import os
import pandas as pd

# Try loading via datasets library
try:
    from datasets import load_dataset
    fiqa = load_dataset("LLukas22/fiqa")
    fiqa["train"].to_pandas().to_csv("data/fiqa_dataset.csv", index=False)
    print("‚úÖ FiQA dataset downloaded via Hugging Face.")
except Exception as e:
    print(f"‚ö†Ô∏è Hugging Face load failed: {e}")
    # Try manual load path
    csv_path = "data/fiqa.csv"
    if os.path.exists(csv_path):
        df = pd.read_csv(csv_path)
        print("‚úÖ FiQA dataset loaded manually from csv.")
        print(df.head())
    else:
        print("‚ùå Manual file not found at:", csv_path)
```
